```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Math': [80, 85, 90, 95, 100],
    'Physics': [82, 88, 91, 97, 99],
    'Chemistry': [78, 83, 88, 90, 94],
    'Biology': [76, 81, 85, 89, 92]
}

df = pd.DataFrame(data)
corr = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title("Pearson Correlation Matrix")
plt.show()
```

## 🔍 Pearson Correlation Analysis

This notebook enhances the data visualization module by demonstrating how to analyze the relationship between multiple features using Pearson correlation.

Pearson correlation measures the strength and direction of a **linear relationship** between two continuous variables. It returns a value between -1 and 1:
- **+1** → Perfect positive relationship  
- **0** → No linear relationship  
- **-1** → Perfect negative relationship

Here, we calculated the correlation matrix for student scores in Math, Physics, Chemistry, and Biology. The resulting **heatmap** helps us visually understand how strongly each subject score is related to the others.

Such correlation analysis is crucial in **feature engineering and preprocessing**, especially before building machine learning models. It helps identify:
- Redundant features
- Strong predictors
- Potential multicollinearity

This is a helpful addition to meaningful visualizations when analyzing real-world datasets.
